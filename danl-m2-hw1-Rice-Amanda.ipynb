#Question 1
How many players had at least some ice time?

*Answer*

888 players had at least some ice time, measured in seconds. Player Ryan Pulock had the least amount of ice time at 237 seconds. Player Drew Doughty had the most amount of ice time at 133,550 seconds.

**Code Used to Answer Question 1:**
- nhl.describe() to generate descriptive statistics that summarize the central tendency, dispersion, and distribution of each variable.
- nhl.describe(include='all') to process string-type variables.
- nhl.sort_values(["TOI"], ascending=True) to sort ice time (seconds) by ascending order.
- nhl.sort_values(["TOI"], ascending=False) to sort ice time (seconds) by descending order.
- nhl.nsmallest(5, 'TOI') and nhl.nlargest(5, 'TOI') to identify the 5 smallest and largest values for ice time.

```python
nhl.describe() #min = 237.00
```

```python
nhl.describe(include='all') #888.00 players have had at least some ice time
```

```python
nhl.sort_values(["TOI"], ascending=True) #sort values, min = 237
```

```python
nhl.sort_values(["TOI"], ascending=False) #sort values, max = 133550
```

```python
nhl.nsmallest(5, 'TOI') #5 smallest amount of ice time
```

```python
nhl.nlargest(5, 'TOI') #5 largest amount of ice time
```

#Question 2
Who is the top scorer in terms of goals?

*Answer*

Sydney Crosby is the top scorer in terms of goals (see below).

**Code Used to Answer Question 2:**
- nhl.describe(include='all') to generate descriptive statistics that summarize the central tendency, dispersion, and distribution of each variable and to process string-type variables.
